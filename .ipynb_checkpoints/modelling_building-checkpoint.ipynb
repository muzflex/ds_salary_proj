```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('eda_data.csv')
```

```python
df.head()
```

### Choose relevant columns

```python
df.columns
```

```python
df_model = df[['Avg Salary', 'Rating', 'Size', 'Type of ownership', 'Industry', 'Sector', 'Revenue', 'Salary Hourly', 'Salary Employer Provided', 'Job State', 'Company Age', 'Python Job', 'R-Studio Job', 'Spark Job', 'AWS Job', 'Excel Job', 'Job Simplifier', 'Seniority', 'Desc Length']]
```

### Get dummy variables

```python
df_dummy = pd.get_dummies(df_model)
```

```python
df_dummy
```

### Train-Test-Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_dummy.drop('Avg Salary', axis = 1)
y = df_dummy['Avg Salary'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=42)
```

## Building Model

### Multiple linear Regression

```python
import statsmodels.api as sm
```

```python
X_sm = X = sm.add_constant(X)
model = sm.OLS(y,X_sm)
model.fit().summary()
```

```python
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.model_selection import cross_val_score
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
np.mean(cross_val_score(lm, X_train, y_train, scoring='neg_mean_absolute_error', cv = 3))
```

### Lasso Regression

```python
lm_l = Lasso()
```

```python
np.mean(cross_val_score(lm_l, X_train, y_train, scoring='neg_mean_absolute_error', cv = 3))
```

```python
alpha = []
error = []

for i in range(1,1000):
    alpha.append(i/1000)
    lm_l = Lasso(alpha=(i/1000))
    error.append(np.mean(cross_val_score(lm, X_train, y_train, scoring='neg_mean_absolute_error', cv = 3)))
    
plt.plot(alpha,error)
```

```python
err = tuple(zip(alpha,error))
df_err = pd.DataFrame(err, columns = ['alpha', 'error'])
```

```python
df_err[df_err.error == max(df_err.error)]
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()

cross_val_score(rf, X_train, y_train, scoring = 'neg_mean_absolute_error', cv=3)
```

```python
np.mean(cross_val_score(rf, X_train, y_train, scoring = 'neg_mean_absolute_error', cv=3))
```

### Tune Model GridsearchCV

```python
from sklearn.model_selection import GridSearchCV
parameters = {'n_estimators': range(10,100,20), 'criterion':('mse','mae'), 'max_features':('auto', 'sqrt','log2')}
```

```python
gs = GridSearchCV(rf,parameters,scoring='neg_mean_absolute_error',cv=3)
```

```python
gs.fit(X_train, y_train)
```

```python
gs.best_score_
```

```python
gs.best_estimator_
```

### Test Ensemble

```python
tpred_lm = lm.predict(X_test)
tpred_lm_l = lm_l.predict(X_test)
tpred_rf = gs.best_estimator_.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test.tpred_lm)
mean_absolute_error(y_test.tpred_lm_l)
mean_absolute_error(y_test.tpred_rf)
```

```python
mean_absolute_error(y_test,(tpred_lm+tpred_rf)/2)
```
